```python
import pandas as pd

# Load the datasets
student_mat = pd.read_csv("dataset2/student-mat.csv", sep=",")
student_por = pd.read_csv("dataset2/student-por.csv", sep=",")

# Use all columns for merging
matching_columns = [
    "school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"
]

# Common students in both datasets (382 guys)
common_students = pd.merge(student_mat, student_por, on=matching_columns)


# Perform an outer merge with an indicator column
all_students = pd.merge(
    student_mat,
    student_por,
    on=[
        "school", "sex", "age", "address", "famsize", "Pstatus", 
        "Medu", "Fedu", "Mjob", "Fjob", "reason", "nursery", "internet"
    ],
    how="outer",
    indicator=True,  # Adds a '_merge' column to indicate the source
    
)

# Save the result to a CSV file for further inspection
all_students.to_csv("all_students_full_dataset.csv", index=False)

# Print the counts for each group
print(all_students["_merge"].value_counts())

# Preview the dataset
print(all_students.head(1000))
```
